```python
import pandas as pd
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
plot.rcParams["figure.figsize"] =(6,6)
```

```python
# define function to visualise the result with matplot
def Visual(X=0, Y=1, test=False):
    plot.scatter(X, Y, color='red', s=7)
    plot.plot(xTrain, linearRegressor.predict(xTrain), linestyle="--", c='#000000')
    if test == True:
        plot.title('Correlation between speed and probability of death (TEST)')
    else:
        plot.title('Correlation between speed and probability of death')
    plot.xlabel('Speed (Km/H)')
    plot.ylabel('% to death')
    plot.show()
```

```python
# import dataset
dataset = pd.read_csv('RoadAccident.csv')
dataset = dataset[dataset['Speed'] < 241]
print(dataset)
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
# split dataset
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size =(35/100), random_state = 0)

# create an fit() linearRegression
linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)

# Predicting the test set results
prediction = linearRegressor.predict(xTest)

# visual of linear regression
Visual(X=xTrain, Y=yTrain)
Visual(X=xTest, Y=yTest, test=True)
```
